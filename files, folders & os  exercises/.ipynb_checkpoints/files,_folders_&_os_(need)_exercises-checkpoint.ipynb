## Exercise 1 (Tidying a Collaboration)

```python
import os
import glob
import shutil
```

```python
for png_file in glob.glob('data/*/*.png', recursive=True):
    os.remove(png_file)
    print(f'{png_file} removed successfully.')
```

```python
for jpg_file in glob.glob('data/*/*.jpg', recursive=True):
    os.remove(jpg_file)
    print(f'{jpg_file} removed successfully.')
```

```python
for pdf_file in glob.glob('data/*/*.pdf', recursive=True):
    os.remove(pdf_file)
    print(f'{pdf_file} removed successfully.')
```

### Task 1

```python
month_to_number={
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
path=os.path.join('data','os-collaboration-exercise-data','oct-08_singapore.txt')
filename=os.path.split(path)[-1]
split1=filename.split('-')
split2=split1[-1].split('_')
filename_parts=[split1[0]]
filename_parts.extend(split2)

for key,value in month_to_number.items():
    if filename_parts[0] == key:
        filename_parts[0] = value

x,y,z = filename_parts
improved_filename=f'{x}-{y}-{z}'
print(improved_filename)
```

### Task 2

```python
def rename_my_file(old_file_name):
    try:
        stringer = str(old_file_name)
        #print(f'You have entered {stringer}.')
        path=os.path.join('data','os-collaboration-exercise-data',stringer)
        if os.path.exists(path) == True:
            #print(f'Your file is located at {path}.')
            filename=os.path.split(path)[-1]
            split1=filename.split('-')
            split2=split1[-1].split('_')
            filename_parts=[split1[0]]
            filename_parts.extend(split2)
            
            for key,value in month_to_number.items():
                if filename_parts[0] == key:
                    filename_parts[0] = value
                
            x,y,z = filename_parts
            improved_filename=f'{x}-{y}-{z}'
            return improved_filename
        else:
            print(f'{old_file_name} does not exist.')
    except:
        print('File already renamed.')

rename_my_file('oct-07_dhaka.txt')
```

### Task 3

```python
entries=['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for entry in entries:
    new_file_name=rename_my_file(entry)
    if new_file_name == None:
        print('Unable to rename.')
    else:
        print(f'{entry} ----> {new_file_name}')
```

### Task 4

```python
for full_filepath in glob.glob(os.path.join('data', '*', '*.txt')):
    text_file=os.path.split(full_filepath)[-1]
    renamed_file=rename_my_file(text_file)
    os.rename(full_filepath, os.path.join('data','os-collaboration-exercise-data',renamed_file))
    print(f'{text_file} ----> {renamed_file}')
```

### Task 5

```python
folder_list=['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']
for folder in folder_list:
    path=os.path.join('data', folder)
    if os.path.exists(path) == False:
        os.mkdir(path)
        print(f'Folder {folder} created.')
    else:
        print(f'Folder {folder} already exists.')
```

### Task 6

```python
glob.glob('data/os-collaboration-exercise-data/*cairo.txt')
```

```python
for cairo_file in glob.glob('data/os-collaboration-exercise-data/*cairo.txt'):
    filename=os.path.split(cairo_file)[-1]
    path=os.path.join('data', 'Cairo', filename)
    shutil.copy(cairo_file, path)
    print(f'Copied {filename} to destination {path}.')
```

### Task 7

```python
for country_name in folder_list:
    country=country_name.lower()
    glob_path=f'data/os-collaboration-exercise-data/*{country}.txt'
    for file in glob.glob(glob_path):
        filename=os.path.split(file)[-1]
        file_path=os.path.join('data', country, filename)
        shutil.copy(file, file_path)
        print(f'Copied {filename} to destination {file_path}.')
```
